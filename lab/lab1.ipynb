```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/umaimehm/Intro_to_AI_2021/main/Lab1/stud.csv"
```

```python
# df = pd.read_csv(url, index_col='StudentID')
df = pd.read_csv(url)
```

```python
df.info()
```

```python
df.sort_index()
```

```python
df.isna().sum()
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True) #Replaces blank values with NaN
```

```python
df.info()
```

```python
df["FinalGrade"].plot()
```

```python
bins = [0, 50, 60, 70, 80, 90, 100]
labels = ['F', 'E', 'D', 'C', 'B', 'A']

df['Grade'] = pd.cut(df['FinalGrade'], bins=bins, labels=labels)
df.head(50)
```

```python
df['Grade'].value_counts().sort_index(ascending=False).plot(kind='bar', xlabel='Grade', ylabel='Count', rot=0)
```
